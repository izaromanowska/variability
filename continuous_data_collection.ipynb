## Continuous collecting data 

```python
from IPython.display import Image
```

We collect data in the cell directly before and directly after a 'barrier'. 

Initially, we've been collecting data once per run, i.e., when the agents crossed the barrier. 

This looked rather random and mirroring the environmental conditions at the time of the arrival at the barrier so we decided to collect it coninuously (at each time step) throughout the simulation run.


The data is collected at each time step and averaged. We started the simulation at 8 different points of the sine curve (in half, quarters, half-quarters).
